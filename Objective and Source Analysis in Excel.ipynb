```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data_set 2.csv")
```

```python
df.head()
```

```python
objective_summary = df['What are your savings objectives?'].value_counts().reset_index()
objective_summary.columns = ['Savings Objective', 'Count']

# Show summary
print("Savings Objectives Summary:")
print(objective_summary)
```

```python
plt.figure(figsize=(7, 7))
colors = plt.cm.Paired.colors  
plt.pie(objective_summary['Count'], 
        labels=objective_summary['Savings Objective'], 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=colors, 
        wedgeprops={'width': 0.4})  

plt.title("Distribution of Savings Objectives")
plt.tight_layout()
plt.savefig("Savings_Objectives_Donut.png")
plt.show()
```

```python
# Count each information source
source_summary = df['Source'].value_counts().reset_index()
source_summary.columns = ['Information Source', 'Count']

# Show summary
print("Information Sources Summary:")
print(source_summary)
```

```python
# Horizontal bar chart for sources
plt.figure(figsize=(9, 6))
plt.barh(source_summary['Information Source'], source_summary['Count'], color="skyblue")
plt.xlabel("Number of Participants")
plt.title("Common Information Sources")
plt.tight_layout()
plt.savefig("Information_Sources_Bar.png")
plt.show()
```

```python
# Export summaries to Excel
with pd.ExcelWriter("Objective_Source_Analysis.xlsx") as writer:
    objective_summary.to_excel(writer, sheet_name='Savings Objectives', index=False)
    source_summary.to_excel(writer, sheet_name='Information Sources', index=False)

print("âœ… Excel file and charts saved successfully.")
```
